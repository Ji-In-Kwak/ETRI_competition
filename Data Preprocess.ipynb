```python
import os
import sys
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('dataset/data/IP/DHCP.csv')
data.head()
```

```python
data.fillna(0, inplace=True)
```

```python
data['Time'] = data['Timestamp'].apply(lambda s: str(s).split('-')[0])
data['Time'] = pd.to_datetime(data['Time'], format='%Y%m%d_%H%M')
```

```python
data.info()
```

```python
data.head()
```

```python
train_data = data[data.Time < datetime(2021, 7, 1)]
train_data
```

```python
test_data = data[data.Time >= datetime(2021, 7, 1)]
test_data
```

```python
train_server = train_data.loc[:, ['Time', 'Svr_detect', 'Svr_connect', 'Ss_request']]
train_server['attack'] = np.random.randint(0,2, size=len(train_server))
train_server.to_csv('GDN-main/data/IP_server/train.csv', index=False)
```

```python
test_server = test_data.loc[:, ['Time', 'Svr_detect', 'Svr_connect', 'Ss_request']]
test_server['attack'] = np.random.randint(0,2, size=len(test_server))
test_server.to_csv('GDN-main/data/IP_server/test.csv', index=False)
```

```python
with open('GDN-main/data/IP_server/list.txt', 'w') as f:
    for col in ['Svr_detect', 'Svr_connect', 'Ss_request']:
        f.write(col)
        f.write('\n')
```

```python
train_server = train_data.loc[:, ['Time', 'Ss_Established']]
train_server['attack'] = np.random.randint(0,2, size=len(train_server))
train_server.to_csv('GDN-main/data/IP_terminal/train.csv', index=False)
```

```python
test_server = test_data.loc[:, ['Time','Ss_Established']]
test_server['attack'] = np.random.randint(0,2, size=len(test_server))
test_server.to_csv('GDN-main/data/IP_terminal/test.csv', index=False)
```
